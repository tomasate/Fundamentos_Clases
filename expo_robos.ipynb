Se empieza una depuracion de la base de datos para quitar informacion que no vamos a usar como objeto de estudio en nuestro analisis. Y se genera un bueno set de datos a analizar.

```python

# Eliminar columnas vacías o con información irrelevante
columnas_a_eliminar = [
    'seguridad.nivel_academico',
    'seguridad.testigo',
    'seguridad.conducta',
    'seguridad.conducta_especial',
    'seguridad.codigo_barrio',
    'seguridad.color'
]
```

```python
# Crear DataFrame con las columnas restantes
df_robos = df.drop(columns=columnas_a_eliminar)

# Limpieza de valores faltantes
df.replace('Sin dato', pd.NA, inplace=True)

# Omitir los valores -1 en la columna 'Edad'
df_robos = df_robos[df_robos['seguridad.edad'] != -1]

# Contar cuántos valores -1 hay en la columna 'Edad'
valores_negativos = df_robos[df_robos['seguridad.edad'] == -1].shape[0]
print(f"Número de valores -1 en la columna 'Edad': {valores_negativos}")

# Codificación de variables
df_robos = pd.get_dummies(df, columns=['seguridad.modalidad'])

df_robos.head()
```

```python
#Limpiar los títulos de las columnas haciendo lo siguiente:

"""
    1. Quitar la palabra "seguridad" al inicio de cada columna
    2. Poner la primera letra de cada título en mayúsculas
    3. Reemplazar la palabra "Bien por "Objeto Robado
    4. Reemplazar la palabra "Peso" por "Dinero"
    5. Reemplazar la expresión "Accerosios prendas de vestir" por "Accerosios/Prendas"
"""

df_robos.columns = df_robos.columns.str.replace(
    'seguridad.', '').str.replace('_', ' ').str.capitalize()
df_robos.columns = df_robos.columns.str.replace(
    r'(?i)bien', 'Objeto Robado', regex=True)
df_robos['Objeto Robado'] = df_robos['Objeto Robado'].replace({
    'Peso': 'Dinero',
    'Accesorios prendas de vestir': 'Accesorios/Prendas'
})
```

# **Organizar la data por grupo etarios**
Se agruparon en tres grupos:

-Niño/joven: Edad de hasta 25 años

-Adulto: Edad comprendida entre los 26 y los 50 años

-Adulto Mayor: Edad mayor a 50 años

A partir de esta definicion de grupos se representan gráficos para comparar la distribucion de edad por objeto robado. Se tienen en cuenta solo el top 10 de objetos robados.

```python
# Crear una función para asignar a cada edad un grupo
'''
Creamos la función "asignar_grupo_edad", tomando una edad como entrada y clasificandola en uno de tres grupos basados en rangos específicos dados.
Si la edad está entre 7 y 25 años, se asigna al grupo "niño/joven". Se define asi puesto que el dato entero menor diferente de cero (0) es el 7.
Si la edad está entre 26 y 50 años, se clasifica en el grupo "adulto".
Para edades mayores a 50 años, se asigna al grupo "adulto mayor".
Esta función se crea para útil para categorizar edades en grupos generacionales con el propósito de análisis y visualización de datos.
'''
def asignar_grupo_edad(edad):
    if 7 <= edad <= 25:
        return 'niño/joven'
    elif 25 < edad <= 50:
        return 'adulto'
    else:
        return 'adulto mayor'

# Aplicar la función y agregar una nueva columna "Grupo Edad" al DataFrame
df_robos['Grupo Edad'] = df_robos['Edad'].apply(asignar_grupo_edad)

# Filtrar el DataFrame solo para el grupo "niño/joven"
df_nino_joven = df_robos[df_robos['Grupo Edad'] == 'niño/joven']

# Calcular el top 10 de objetos robados más comunes en el grupo "niño/joven"
top10_objetos_robados_nino_joven = df_nino_joven['Objeto Robado'].value_counts().head(10).index

# Filtrar el DataFrame del grupo "niño/joven" solo para los top 10 objetos robados
df_top10_nino_joven = df_nino_joven[df_nino_joven['Objeto Robado'].isin(top10_objetos_robados_nino_joven)]

# Gráfico de caja sin valores atípicos para comparar la distribución de edad por objeto robado (top 10) para el grupo "niño/joven"
plt.figure(figsize=(35, 10))
ax = sns.boxplot(data=df_top10_nino_joven, x='Objeto Robado', y='Edad', showfliers=False)  # Ocultar outliers
plt.xlabel('Objeto Robado')
plt.ylabel('Edad')
plt.title('Distribución de Edad por Objeto Robado (Top 10) para el Grupo "Niño/Joven"')
plt.xticks(rotation=45, ha='right')

# Ajustar las etiquetas del eje x para que sean solo para el top 10
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

# Agregar etiquetas con mediana y desviación estándar
medians = df_top10_nino_joven.groupby('Objeto Robado')['Edad'].median()
stds = df_top10_nino_joven.groupby('Objeto Robado')['Edad'].std()
for xtick, label in zip(ax.get_xticks(), ax.get_xticklabels()):
    objeto_robado = label.get_text()
    median_value = medians[objeto_robado]
    std_value = stds[objeto_robado]
    ax.text(xtick, median_value, f'Median: {median_value:.1f}\nSD: {std_value:.1f}', ha='center', color='black')

plt.show()
```

**Determinar si hay una asociación entre género y número de registros :**

```python
from scipy.stats import chi2_contingency

# Crear una tabla de contingencia entre género y el número total de registros
tabla_contingencia = pd.crosstab(df_robos['Sexo'], df_robos['Sexo'])

# Realizar el test de Chi-cuadrado
chi2, p, dof, expected = chi2_contingency(tabla_contingencia)

# Comparar el valor de p con un umbral (por ejemplo, 0.05) para determinar la significancia
alpha = 0.05
if p < alpha:
    print("Hay una asociación significativa entre género y número de registros.")
else:
    print("No hay una asociación significativa entre género y número de registros.")
```

# **Pruebas de Hipótesis:**
**1. Prueba de Hipótesis para el Número de Robos por Género:**
Se hará una prueba de Chi-Cuadrado con las siguientes hipótesis:

$H 0$=  No existe asociación entre el género y si una persona fue robada

$H 1$= Existe alta relación entre el género y si una persona fue robada

Usando la fórmula
$χ2=∑Eij(Oij−Eij)2$

```python
df_robos['Fue Robado'] = df_robos['Objeto Robado'].notnull().astype(int)

from scipy.stats import chi2_contingency

# Crear una tabla de contingencia entre género y si fue robado (0 si no fue robado, 1 si fue robado)
tabla_contingencia = pd.crosstab(df_robos['Sexo'], df_robos['Fue Robado'])

# Realizar el test de chi-cuadrado para evaluar la asociación
chi2, p_valor, dof, expected = chi2_contingency(tabla_contingencia)

print("Prueba Chi-cuadrado para la Asociación entre Género y si Fue Robado:")
print("Estadístico Chi-cuadrado:", chi2)
print("Valor P:", p_valor)
```

El valor p obtenido en la prueba de Chi-cuadrado es muy pequeño $(7.15e-07)$, lo que indica que existe una asociación significativa entre el género y si una persona fue robada. En otras palabras, el género parece tener influencia en la probabilidad de que una persona haya sido robada.

Dado que el valor p es mucho menor que el umbral de significancia comúnmente utilizado (como $0.05$), podemos rechazar la hipótesis nula, que establece que no hay asociación entre las variables. En su lugar, podemos concluir que hay una asociación significativa entre el género y si una persona fue robada, según los datos que tienes disponibles.
