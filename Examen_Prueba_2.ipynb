# **Solución**
### **Ejercicio 2**

**Paso 2.1: Carga de Archivo y Renombre de Dataset**

Empezamos cargando un archivo CSV desde una URL y asignándolo a un DataFrame. Además, renombramos el dataset por simplicidad.

```python
# Carga de archivo y renombre de dataset#

import pandas as pd

# URL del archivo CSV
url = "https://github.com/tomasate/Datos_Clases/raw/main/Datos_1/athletes.csv"

# Leer el archivo CSV desde la URL
athletes = pd.read_csv(url)

athletes
```

**Paso 2.2: Filtrar Nombres con Medallas**

En esta parte, filtramos los atletas que han ganado al menos una medalla y creamos un nuevo DataFrame para estos atletas.

```python
# Filtrar nombres con medallas.

# Calcula el total de medallas ganadas por cada atleta y evalúa si es mayor a 0
athletes_with_medals = ((athletes['gold'] + athletes['silver'] + athletes['bronze']) > 0)
athletes_filtered = athletes[athletes_with_medals]

# Crea el DataFrame 'winner_athletes' con los atletas que tienen medallas mayores a 0
winner_athletes = pd.DataFrame({'Athlete': athletes_filtered['name']})

# Agrega una nueva columna 'Total Medals' al DataFrame 'winner_athletes' con el número total de medallas
winner_athletes['Total Medals'] = athletes_filtered['gold'] + athletes_filtered['silver'] + athletes_filtered['bronze']

winner_athletes
```

**Paso 2.3: Calcular Indice de Masa Muscular**

Aquí calculamos la masa muscular de los atletas utilizando las columnas de peso y altura o IMC.

```python
# Calcular Indice de Masa Muscular:

'''
Se define función para el cálculo de la masa corporal de los atletas por fila,
tomando en cuenta las columnas de peso y altura o el indice de masa corporal (IMC).
'''

def calculate_muscle_mass(row):
    return row['weight'] / (row['height'] ** 2)

# Aplicar la función a cada fila y crear una nueva columna 'Muscle Mass'
winner_athletes['Muscle Mass'] = athletes.apply(calculate_muscle_mass, axis=1)

winner_athletes
```

**Paso 2.4: Mayor Peso Promedio por Deporte**

En este paso, solicitamos al usuario la columna por la cual desea ordenar y luego agrupamos los datos por deporte para calcular el peso promedio, altura promedio y promedio de medallas por deporte.

```python
# Mayor Peso Promedio

# Solicitar al usuario la columna por la cual desea ordenar
column_to_sort = input("Ingrese el nombre de la columna por la cual desea ordenar: ")

# Agrupar los datos por deporte y calcular mean por cada una.
grouped_by_sport = pd.DataFrame(athletes.groupby('sport').agg({
    'height': 'mean',
    'weight': 'mean',
    'gold': 'mean',
    'silver': 'mean',
    'bronze': 'mean'
}))

# Ordenarlos por columna elegida.
grouped_by_sport.sort_values(by=column_to_sort, ascending=False, inplace=True)

grouped_by_sport
```

**Paso 2.5: Encontrar el País con Más Participación Femenina**

En esta sección, filtramos los países por género femenino y encontramos el país con la mayor participación de atletas femeninas.

```python
# 5. Encontrar el país con más participación femenina

# Se filtra países por género femenino, teniendo encuenta el máximo conteo.
country_with_most_female_participation = athletes[athletes['sex'] == 'female']['nationality'].value_counts().idxmax()

# Se imprime país con mayor participación femenina.
print(country_with_most_female_participation)
```

Importe el dataset de nba del repositorio de clase y responda (hint: Elimine las instancias con registros faltantes):

# **Solución**
### **Ejercicio 3**

**Paso 1.1: Importación y Carga de Datos**

En el primer paso, nos aseguramos de tener la librería Pandas importada y cargamos un conjunto de datos de la NBA desde una fuente en línea.

```python
#Nos aseguramos que pandas este importado
import pandas as pd

# Importamos la Base de Datos
df = pd.read_csv("https://raw.githubusercontent.com/tomasate/Datos_Clases/main/Datos_1/nba.csv")
df
```

```python
#num_teams = len(df['Team'].unique())
#print("Número de equipos:", num_teams)
```

**Paso 1.2: Limpieza de Datos - Eliminación de Valores Faltantes**

La limpieza de datos es crucial para asegurar la calidad de nuestro análisis. Aquí eliminamos registros con valores faltantes del conjunto de datos.

```python
# Eliminamos registros con valores faltantes de la Base de Datos
df.dropna(inplace=True)
df
```

**Paso 1.3: Equipos con los Salarios Más Altos**

En esta sección, determinamos: ¿Cuáles son los cinco equipos que pagan los salarios más altos?

```python
# Generamos una funcion para determinar los 5 equipos que pagan los salarios más altos
'''
La funcion permite realizar una agrupacion de los equipos y ordenarlos de de forma ascendente
de acuerdo a los salarios reportados, y en este punto mostrar los 5 primeros.
'''

top_salaries = df.groupby('Team')['Salary'].sum().sort_values(ascending=False).reset_index().head(5)
top_salaries.columns = ['Equipo', 'Salario']
#top_salaries.index = [1,2,3,4,5]
top_salaries.index = [i for i in range(1,6)]
top_salaries
```

**Paso 2.1: Crear Categorías de Edad**
En este paso, estamos dividiendo la variable 'Age' en grupos y agregando una nueva columna 'Age-Group' al DataFrame para categorizar a los jugadores en grupos de edad.

```python
# Separamos la variable Age en grupos
bins = [0, 25, 30, df['Age'].max()]

# Creamos  unas etiquetas "labels" y la columna Age-Group
labels = ['Jóvenes', 'Intermedios', 'Mayores']
df['Age-Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
# Visualizamos
df
```

**Paso 3.1: Crear Tabla Dinámica de Salarios por Equipo y Posición**

En este paso, estamos filtrando el DataFrame original por los equipos en 'top_salaries', luego creamos una tabla dinámica que muestra el salario promedio de los jugadores por equipo y posición.

```python
# Pregunta 3
# Filtramos el DataFrame original por los equipos en top_salaries
filtered_df = df[df['Team'].isin(top_salaries['Equipo'])]

# Creamos la tabla dinámica
pivot_table = pd.pivot_table(filtered_df,
                             values='Salary',
                             index=['Team', 'Position'],
                             aggfunc=lambda x: '{:,.0f}'.format(x.mean()))

print(pivot_table)
```

Los resultados reflejan los salarios promedio de los jugadores en diferentes equipos y posiciones. Se observa que hay diferencias notables en los salarios según la posición de juego y el equipo. Por ejemplo, en equipos como _Cleveland Cavaliers_ y _Golden State Warriors_, los jugadores en la posición de Power Forward (PF) tienden a tener salarios más altos. Los datos también sugieren que los equipos tienen enfoques diversos en la distribución de salarios, con algunas posiciones ganando más que otras en distintos equipos. Esto podría reflejar estrategias individuales de juego por equipo, que impactan directamente en la gestión de salarios.
